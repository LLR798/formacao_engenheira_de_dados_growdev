##Desafio listas e funções:

Crie uma função que representa o jogo batalha naval.

Dentro da função crie uma estrutura bidimensional utilizando listas com sublistas para representar um tabuleiro (1 lista com 20 elementos e cada elemento é uma lista de 20 elementos, tabuleiro 20x20). Cada posição irá armazenar 1 valor numérico que significa:
- 0 - Água
- 1 - Navio

Para cada posição escolha esses valores aleatoriamente, respeitando a regra de que não podem existir mais de 20 navios no tabuleiro.

Após os valores serem distribuídos, o programa deve pedir ao usuário uma posição do tabuleiro e informar se ele acertou um navio ou água e repetir o procedimento até que o usuário derrote todos os navios ou chegue ao limite de 35 tentativas.



```python
import random

navios_colocados = 0

erros = 0
tentativas = 0
navios_atingidos = 0

# Criar o tabuleiro vazio
tabuleiro = [[0] * 20 for _ in range(20)]

# Função para imprimir o tabuleiro
def imprimir_tabuleiro():
    for linha in tabuleiro:
        print(" ".join(str(valor) for valor in linha))
    print()

# Distribuir os navios aleatoriamente
while navios_colocados < 20:
    
    x = random.randint(0, 19)
    y = random.randint(0, 19)
    
    if tabuleiro[x][y] == 0:
        tabuleiro[x][y] = 1
        navios_colocados += 1

while tentativas < 35 and navios_atingidos < 20:
    print(f"Tentativa {tentativas + 1}")
    
    x = int(input("Digite a linha (0-19): "))
    y = int(input("Digite a coluna (0-19): "))
    
    if 0 <= x < 20 and 0 <= y < 20:
        if tabuleiro[x][y] == 1:
            print("Parabéns! Você acertou um navio!")
            tabuleiro[x][y] = '1'
            navios_atingidos += 1
        else:
            print("Água. Tente novamente.")
            # tabuleiro[x][y] = '0' 
            erros += 1
    else:
        print("Posição inválida. Tente novamente.")
    imprimir_tabuleiro()
    tentativas += 1

# Verificar se o jogador ganhou ou perdeu
if navios_atingidos == 20:
    print("Parabéns! Você derrotou todos os navios!")
else:
    print(f"Fim do jogo! Você acertou {navios_atingidos} navios, errou {erros} vezes, e usou {tentativas} tentativas.")
```

- Acertos em água

```python
import random

navios_colocados = 0
tentativas = 0
navios_atingidos = 0
acertos_agua = 0
acertos_navios = 0
acertos_ininterruptos = 0
acertos_atuais = 0

# Criar o tabuleiro vazio
tabuleiro = [[0] * 20 for _ in range(20)]

# Função para imprimir o tabuleiro
def imprimir_tabuleiro():
    for linha in tabuleiro:
        print(" ".join(str(valor) for valor in linha))
    print()

# Distribuir os navios aleatoriamente
while navios_colocados < 20:
    x = random.randint(0, 19)
    y = random.randint(0, 19)
    if tabuleiro[x][y] == 0:
        tabuleiro[x][y] = 1
        navios_colocados += 1

# Loop principal do jogo
while tentativas < 35 and navios_atingidos < 20:
    print(f"Tentativa {tentativas + 1}")
    x = int(input("Digite a linha (0-19): "))
    y = int(input("Digite a coluna (0-19): "))

    if 0 <= x < 20 and 0 <= y < 20:
        if tabuleiro[x][y] == 1:
            print("Parabéns! Você acertou um navio!")
            tabuleiro[x][y] = '1'  # Marca o navio atingido
            navios_atingidos += 1
            acertos_navios += 1
            acertos_atuais += 1
        else:
            print("Água. Tente novamente.")
            acertos_agua += 1
            acertos_atuais = 0  # Reinicia a sequência de acertos
    else:
        print("Posição inválida. Tente novamente.")

    # Atualiza a maior sequência de acertos
    acertos_ininterruptos = max(acertos_ininterruptos, acertos_atuais)

    imprimir_tabuleiro()
    tentativas += 1

# Calcula as porcentagens de acerto
porcentagem_agua = (acertos_agua / tentativas) * 100 if tentativas > 0 else 0
porcentagem_navios = (acertos_navios / tentativas) * 100 if tentativas > 0 else 0

# Resultado final
if navios_atingidos == 20:
    print("Parabéns! Você derrotou todos os navios!")
else:
    print(f"Fim do jogo! Você acertou {navios_atingidos} navios.")

# Exibe as estatísticas
print("\nEstatísticas:")
print(f"Acertos em água: {acertos_agua}")
print(f"Acertos em navios: {acertos_navios}")
print(f"Porcentagem de acertos em água: {porcentagem_agua:.2f}%")
print(f"Porcentagem de acertos em navios: {porcentagem_navios:.2f}%")
print(f"Maior número de acertos ininterruptos: {acertos_ininterruptos}")
```

```python
import csv

def ler_csv_para_dicionario(caminho_arquivo):
    dados = {}
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor = csv.reader(csvfile)
        cabecalho = next(leitor)  # Lê a linha de cabeçalho

        for i, linha in enumerate(leitor):
            individuo = {cabecalho[j]: linha[j] for j in range(len(cabecalho))}
            dados[i] = individuo

    return dados
```

Busque quais são os anos da base de dados?

```python
import csv

def extrair_anos_unicos(nome_arquivo=diretorio, nome_coluna="ano"):
    anos = set() 

    with open(nome_arquivo, 'r', newline='') as arquivo_csv:
        leitor_csv = csv.DictReader(arquivo_csv)  
        for linha in leitor_csv:
            try:
                ano = int(linha[nome_coluna]) 
                anos.add(ano) 
            except (ValueError, KeyError):
                pass  

    return sorted(list(anos))  

anos_unicos = extrair_anos_unicos()
print(anos_unicos)
```

Qual a porcentagem de homens e mulheres na base de dados?

```python
import csv

def calcular_porcentagem_sexos(nome_arquivo=diretorio, nome_coluna="sexo"):
    contagem = {"M": 0, "F": 0} 
    total_linhas_validas = 0

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                sexo = linha.get(nome_coluna)  
                if sexo in contagem:  
                    contagem[sexo] += 1
                    total_linhas_validas += 1
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return {"Masculino": 0, "Feminino": 0}

    porcentagens = {}
    for sexo, quantidade in contagem.items():
        porcentagem = (quantidade / total_linhas_validas) * 100 if total_linhas_validas > 0 else 0
        porcentagens[sexo if sexo == "M" else "Feminino"] = round(porcentagem, 2) 

    return porcentagens

porcentagens_sexo = calcular_porcentagem_sexos()
print(porcentagens_sexo)
```

Qual foi o gasto por ano?

```python
import csv

def calcular_gasto_por_ano(nome_arquivo=diretorio, nome_coluna_ano="ano", nome_coluna_valor="compra"):
    gastos_por_ano = {}

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    ano = int(linha[nome_coluna_ano])
                    valor = float(linha[nome_coluna_valor])
                    gastos_por_ano[ano] = gastos_por_ano.get(ano, 0) + valor
                except (ValueError, KeyError):
                    pass 
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return {} 

    return gastos_por_ano

gastos = calcular_gasto_por_ano()
print(gastos)
```

Qual foi o ano com maior gasto?

```python
def encontrar_ano_maior_gasto(gastos_por_ano):
    if not gastos_por_ano:
        return None, None

    ano_maior_gasto = max(gastos_por_ano, key=gastos_por_ano.get)
    maior_gasto = gastos_por_ano[ano_maior_gasto]
    return ano_maior_gasto, maior_gasto

gastos = calcular_gasto_por_ano()

ano_maior, valor_maior = encontrar_ano_maior_gasto(gastos)
if ano_maior is not None:
    print(f"O ano com maior gasto foi {ano_maior}, com um total de R$ {valor_maior:.2f}")
else:
    print("Não foi possível calcular os gastos. Verifique o arquivo e tente novamente.")
```

Utilizando as faixas etárias, diga quantas pessoas há em cada faixa?

```python
import csv

def contar_pessoas_por_faixa_etaria(nome_arquivo=diretorio, nome_coluna_idade="idade", jovem_max=25, adulto_max=59):
    contagem_faixas = {'Jovens': 0, 'Adultos': 0, 'Idosos': 0}

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    idade = int(linha[nome_coluna_idade])
                    if idade <= jovem_max:
                        contagem_faixas['Jovens'] += 1
                    elif idade <= adulto_max:
                        contagem_faixas['Adultos'] += 1
                    else:
                        contagem_faixas['Idosos'] += 1
                except (ValueError, KeyError):
                    pass  
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return {}

    return contagem_faixas

contagem = contar_pessoas_por_faixa_etaria()
print(contagem)
```

Qual é a faixa etária que mais gasta?

```python
import csv

def faixa_etaria_que_mais_gasta(nome_arquivo=diretorio, nome_coluna_idade="idade", nome_coluna_valor="compra", jovem_max=25, adulto_max=59):
    gastos_por_faixa = {'Jovens': 0, 'Adultos': 0, 'Idosos': 0}

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    idade = int(linha[nome_coluna_idade])
                    valor = float(linha[nome_coluna_valor])

                    if idade <= jovem_max:
                        gastos_por_faixa['Jovens'] += valor
                    elif idade <= adulto_max:
                        gastos_por_faixa['Adultos'] += valor
                    else:
                        gastos_por_faixa['Idosos'] += valor
                except (ValueError, KeyError):
                    pass  
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None  

    faixa_maior_gasto = max(gastos_por_faixa, key=gastos_por_faixa.get)
    return faixa_maior_gasto

faixa_etaria = faixa_etaria_que_mais_gasta()
if faixa_etaria:
    print(f"A faixa etária que mais gasta é: {faixa_etaria}")
```

A opção de débito é mais utilizada entre homens ou mulheres?

```python
import csv

def debito_mais_usado_por_sexo(nome_arquivo=diretorio, nome_coluna_sexo="sexo", nome_coluna_pagamento="pagamento"):
    contagem_pagamentos = {"Masculino": 0, "Feminino": 0}

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    sexo = linha[nome_coluna_sexo]
                    pagamento = linha[nome_coluna_pagamento]

                    if pagamento == "débito" and sexo in contagem_pagamentos:
                        contagem_pagamentos[sexo] += 1
                except KeyError:
                    pass 
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

    sexo_com_mais_debitos = max(contagem_pagamentos, key=contagem_pagamentos.get)
    if contagem_pagamentos["Masculino"] == contagem_pagamentos["Feminino"]:
        return "Igual entre ambos"
    else:
        return sexo_com_mais_debitos

resultado = debito_mais_usado_por_sexo()
if resultado:
    print(f"O sexo que mais utiliza débito é: {resultado}")
```

Qual o sobrenome que mais aparece na base de dados?

```python
import csv
from collections import Counter

def sobrenome_mais_frequente(nome_arquivo=diretorio, nome_coluna_sobrenome="sobrenome"):
    sobrenomes = []
    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    sobrenome = linha[nome_coluna_sobrenome]
                    sobrenomes.append(sobrenome)
                except KeyError:
                    pass 
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

    contagem_sobrenomes = Counter(sobrenomes)

    sobrenome_mais_frequente = contagem_sobrenomes.most_common(1)[0][0]
    return sobrenome_mais_frequente

resultado = sobrenome_mais_frequente()
if resultado:
    print("O sobrenome mais frequente é:", resultado)
```

Qual foi o ano em que os homens mais usaram o crédito?

```python
import csv
from collections import defaultdict

def ano_mais_credito_homens(nome_arquivo=diretorio, nome_coluna_sexo="sexo", nome_coluna_ano="ano", nome_coluna_pagamento="pagamento"):
    contagem_credito_por_ano = defaultdict(int)  
    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    sexo = linha[nome_coluna_sexo]
                    ano = int(linha[nome_coluna_ano])
                    pagamento = linha[nome_coluna_pagamento]

                    if sexo == "M" and pagamento == "credito":
                        contagem_credito_por_ano[ano] += 1
                except (KeyError, ValueError):
                    pass  
    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

    ano_maximo = max(contagem_credito_por_ano, key=contagem_credito_por_ano.get, default=None)
    return ano_maximo

resultado = ano_mais_credito_homens()

if resultado:
    print(f"O ano em que os homens mais usaram crédito foi: {resultado}")
```

Qual opção de pagamento é mais utilizada em cada faixa etária?

```python
import csv
from collections import defaultdict

def pagamento_mais_usado_por_faixa_etaria(nome_arquivo=diretorio, nome_coluna_idade="idade", nome_coluna_pagamento="pagamento"):
    contagem_pagamentos = {
        "Jovens": defaultdict(int),
        "Adultos": defaultdict(int),
        "Idosos": defaultdict(int)
    }

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    idade = int(linha[nome_coluna_idade])
                    pagamento = linha[nome_coluna_pagamento].strip().lower()

                    if 18 <= idade <= 25:
                        contagem_pagamentos["Jovens"][pagamento] += 1
                    elif 26 <= idade <= 59:
                        contagem_pagamentos["Adultos"][pagamento] += 1
                    elif idade >= 60:
                        contagem_pagamentos["Idosos"][pagamento] += 1
                except (KeyError, ValueError):
                    continue 

        resultado = {}
        for faixa_etaria, pagamentos in contagem_pagamentos.items():
            if pagamentos:  
                pagamento_mais_usado = max(pagamentos, key=pagamentos.get)
                resultado[faixa_etaria] = pagamento_mais_usado
            else:
                resultado[faixa_etaria] = None  

        return resultado

    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

resultado = pagamento_mais_usado_por_faixa_etaria()
if resultado:
    for faixa_etaria, pagamento in resultado.items():
        if pagamento:
            print(f"A opção de pagamento mais usada por {faixa_etaria} é: {pagamento}")
        else:
            print(f"Não foram encontrados dados de compras para {faixa_etaria}.")
```

Qual o valor gasto em compras por jovens do ano de 2010 até 2015?

```python
import csv

def valor_gasto_jovens(nome_arquivo=diretorio, nome_coluna_idade="idade", nome_coluna_compra="compra", nome_coluna_ano="ano", ano_inicial=None, ano_final=None):
    total_gasto = 0

    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.DictReader(arquivo_csv)
            for linha in leitor_csv:
                try:
                    idade = int(linha[nome_coluna_idade])
                    valor_compra = float(linha[nome_coluna_compra])
                    ano_compra = int(linha[nome_coluna_ano])

                    if 18 <= idade <= 25 and (ano_inicial is None or ano_compra >= ano_inicial) and (ano_final is None or ano_compra <= ano_final):
                        total_gasto += valor_compra
                except (KeyError, ValueError):
                    continue 

        return total_gasto

    except FileNotFoundError:
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None

valor_total = valor_gasto_jovens(ano_inicial=2010, ano_final=2015) 
if valor_total is not None:
    print(f"O valor total gasto por jovens é: R$ {valor_total:.2f}")
```
